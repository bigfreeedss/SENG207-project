```python
#6 different graphs on the dataset 
# Setting up a 3x2 subplot grid
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 15))

# Plot 1: Histogram of Alcohol Content
axes[0, 0].hist(dataset['alcohol'], bins=20, edgecolor='black', color='skyblue')
axes[0, 0].set_title('1: Histogram of Alcohol Content')
axes[0, 0].set_xlabel('Alcohol Content')
axes[0, 0].set_ylabel('Frequency')

# Plot 2: Scatter Plot of Fixed Acidity vs. Citric Acid
axes[0, 1].scatter(dataset['fixed acidity'], dataset['citric acid'], edgecolor='black', color='coral', alpha=0.6)
axes[0, 1].set_title('2: Fixed Acidity vs. Citric Acid')
axes[0, 1].set_xlabel('Fixed Acidity')
axes[0, 1].set_ylabel('Citric Acid')


# Plot 3: Boxplot of Residual Sugar by Wine Quality
axes[1, 0].boxplot([dataset[dataset['quality'] == i]['residual sugar'] for i in sorted(dataset['quality'].unique())],
                  labels=sorted(dataset['quality'].unique()), patch_artist=True)
axes[1, 0].set_title('3: Boxplot of Residual Sugar by Wine Quality')
axes[1, 0].set_xlabel('Wine Quality')
axes[1, 0].set_ylabel('Residual Sugar')

# Plot 4: Line Plot of pH over Index
axes[1, 1].plot(dataset['pH'])
axes[1, 1].set_title('4: Line Plot of pH over Index')
axes[1, 1].set_xlabel('Index')
axes[1, 1].set_ylabel('pH')

# Plot 5: Bar Chart of Wine Quality Counts
quality_counts = dataset['quality'].value_counts().sort_index()
axes[2, 0].bar(quality_counts.index, quality_counts.values, color='green')
axes[2, 0].set_title('5: Bar Chart of Wine Quality Counts')
axes[2, 0].set_xlabel('Wine Quality')
axes[2, 0].set_ylabel('Count')

# Plot 6: Pie Chart of Sulphates Distribution
sulphates_distribution = dataset['sulphates'].value_counts()
axes[2, 1].pie(sulphates_distribution, labels=sulphates_distribution.index, autopct='%1.1f%%', startangle=90)
axes[2, 1].set_title('6: Pie Chart of Sulphates Distribution')

# Adjust layout
plt.tight_layout()
plt.show()
```

# A REPORT ON THE GRAPHS ILLUSTRATED ABOVE
---

### 1. Histogram of Alcohol Content:
- This histogram illustrates the distribution of alcohol content in the red wine dataset. Each bar represents a range of alcohol content values, and the height of the bars indicates the frequency of wines falling within each range.
- The histogram provides insights into the prevalent alcohol content levels in the dataset and their distribution.



### 2. Scatter Plot of Fixed Acidity vs. Citric Acid
- This scatter plot displays the relationship between fixed acidity and citric acid in red wines. Each point represents a wine, with its position determined by the respective values of fixed acidity and citric acid.
-  The scatter plot helps identify patterns or trends between these two chemical properties. Clusters or trends in the data points may reveal associations between fixed acidity and citric acid.


### 3. Boxplot of Residual Sugar by Wine Quality
- This boxplot provides a visual summary of the distribution of residual sugar for different quality levels of red wines. Each box represents a wine quality category, showing the median, quartiles, and potential outliers.
* The boxplot allows for the comparison of residual sugar distributions across different wine qualities, offering insights into how this chemical property varies with wine quality.

### 4. Line Plot of pH over Index
- This line plot depicts the variation of pH values over the index (or order) of the red wine dataset. Each point on the line corresponds to a specific entry in the dataset, with the x-axis representing the index and the y-axis representing pH values.
- The line plot can reveal trends or patterns in the pH values over the dataset's index, providing a sense of how pH values change sequentially.


### 5. Bar Chart of Wine Quality Counts
- This bar chart displays the counts of each wine quality category in the dataset. Each bar represents a wine quality level, and the height of the bars corresponds to the number of wines in each category.
- The bar chart provides a clear visual representation of the distribution of wine qualities in the dataset, indicating the prevalence of each quality level.

### 6. Pie Chart of Sulphates Distribution
- This pie chart illustrates the distribution of sulphates in the red wine dataset. Each slice of the pie represents a unique sulphate value, with the size of the slice proportional to the frequency of that sulphate value in the dataset.
- The pie chart offers a concise overview of the distribution of sulphates, highlighting the most common sulphate values in the dataset.
